# 1. Import Pandas

```python
# prepare the work station
# import the necessary library
import pandas as pd
```

# 2. View the excel file

```python
# Read the file
movies = pd.read_excel('movies_merge.xlsx')
# view the dataframe
print(movies.columns)
```

# 3. View the csv file

```python
# read the second file
ott = pd.read_csv('ott_merge.csv')
# view the dataframe
print(ott.columns)
```

# 4. Validate the DataFrame

```python
print(movies.shape)
movies.head()
```

```python
print(ott.shape)
ott.head()
```

# 5. Describe the Dataframe

```python
# Determine the data types
print(movies.dtypes)
print(ott.dtypes)
```

# 6. Combine the Dataframe

# a. using merge()

```python
# combine using merge()
movies_ott_combine = pd.merge(movies, ott, how = 'left', on = 'ID')
# View the dataframe
print(movies_ott_combine.shape)
movies_ott_combine.head()
```

# b. using concat()

```python
# combine using concat
movies_ott_conc = pd.concat([movies, ott], axis = 0)
# view the dataframe
print(movies_ott_conc.shape)
movies_ott_conc.head()
```
